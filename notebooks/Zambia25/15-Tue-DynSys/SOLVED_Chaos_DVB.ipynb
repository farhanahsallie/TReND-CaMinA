## **Chaos**

**Chaos refers to the fact very small differences in the initial state of a dynamical system can result in large differences in the trajectories of that dynamical system.**

The discovery that some dynamical systems can be chaotic, had profound consequences to science and philosophy. It meant, that even though a system is deterministically governed by a set of equations, that in practice, it may not be possible to predict the future dynamics of that system. 

The discrete logistic equation ("map") shows that even very simple equations can generate chaos.

$$ x(t) = r*x(t-1)*(1-x(t-1)) $$


$r$ is the "growth", "control", or "gain" parameter. Values above 3.56995 lead to chaos.

First see how two different simulations with very small differences in initial parameters diverge.
The first point of the first simulation starts with x[0,0] = 0.5;
The first point of the second simpulation starts with x[1,0] = 0.50000001

Explore the effects of different initial conditions and r values.


```python
import numpy as np
import matplotlib.pyplot as plt

numsteps = 250
# r = "growth rate" or "gain"
# r>3.56995 results in chaos
r = 3.8      

#note that we are running two simulations in parallel x[0,:] and x[1,:]
x = np.zeros((2,numsteps))
x[0,0] = 0.5
x[1,0] = 0.50000001

for t in range(1,numsteps):
    # the actual equation
    x[:,t] = r*x[:,t-1]*(1-x[:,t-1])

plt.plot(x.T)
plt.show()

plt.plot(x[0,:]-x[1,:])
plt.show()
```


We can plot the consecutive points in the time series (the values of x[0,t], x[0,t+1]) against eachother in a type of a phase plot (a Poincare plot).
Note that even though the system is chaotic, the points always fall along a parabola. So we cannot predict what any future point will be (without running the simulation) but we can predict the region where all points will fall.


```python
plt.scatter(x[0,:-1],x[0,1:],s=6, color='red')  #First simpulation
plt.scatter(x[1,:-1],x[1,1:],s=3, color='blue') #Second simpulation
plt.xlabel('t')
plt.ylabel('t+1')

```
